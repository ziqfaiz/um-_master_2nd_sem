```python
import pandas as pd
```

```python
df = pd.read_csv("salary.csv")
df.head()
```

```python
df["Timestamp"]= pd.to_datetime(df["Timestamp"],format='mixed')
df["Salary"] = pd.to_numeric(df["Salary"].str.replace(",",""))
```

```python

df["Age"].unique()
```

```python
age_order = ['under 18', '18-24', '25-34', '35-44', '45-54', '55-64', '65 or over']

df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)
```

```python
experience_order = sorted(df["Overall years of experience"].unique())
df["Overall years of experience"]= pd.Categorical(df["Overall years of experience"],categories=experience_order,ordered=True)
```

```python
current_experience_order = sorted(df["Years of experience in current job"].unique())
df["Years of experience in current job"]= pd.Categorical(df["Years of experience in current job"],categories=current_experience_order,ordered=True)
```
